# Model Training

```python
from sklearn.model_selection import train_test_split
# train - 70
# test - 30
X = df.drop(columns=['Species'])
Y = df['Species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

## LogisticRegression

```python
# logistic regression 
from sklearn.linear_model import LogisticRegression
LGR = LogisticRegression()
```

```python
# model training
LGR.fit(x_train, y_train)
```

### Model Prediction

```python
prect = LGR.predict(x_test)
```

```python
prect
```

### Accuracy

```python
# print metric to get performance
print("Accuracy: ",LGR.score(x_test, y_test) * 100)
```

## K-Nearest Neighbour

```python
# knn - k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
```

```python
KNN.fit(x_train, y_train)
```

### Model Prediction 

```python
prect = KNN.predict(x_test)
```

```python
prect
```

### Accuracy

```python
# print metric to get performance
print("Accuracy: ",KNN.score(x_test, y_test) * 100)
```

## DecisionTree 

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(x_train, y_train)
```

### Model Prediction

```python
prect = DTC.predict(x_test)
```

```python
prect
```

### Accuracy 

```python
# print metric to get performance
print("Accuracy: ",DTC.score(x_test, y_test) * 100)
```

## SVC

```python
from sklearn.svm import SVC
SVN = SVC()
```

```python
SVN.fit(x_train, y_train)
```

### Model Prediction

```python
prect = SVN.predict(x_test)
```

```python
prect 
```

### Acurracy

```python
print("Accuracy: ",SVN.score(x_test, y_test) * 100)
```
